```python
import os
import requests
import polars as pl
```

```python
df = pl.read_parquet('data_raw/papezstvo_raw.parquet', use_pyarrow=True)
```

```python
seznam = df.select(pl.col(['id','wiki_en'])).rows_by_key(key='id')
seznam = {kod: link[0] for kod, link in seznam.items()}
seznam
```

```python
kam = "downloads/wikipedia_en"
```

```python
try:
    os.mkdir(kam)
except:
    pass
```

```python
for kod, link in seznam.items():
    print(link)
    r = requests.get(link)
    with open(os.path.join(kam, f'{kod}.html'), 'w+', encoding='utf-8') as soubor:
        soubor.write(r.text)
```
