```python
df_vatikan = pl.read_json("data_raw/papezstvo_raw_vatikan2.json").with_columns(pl.col("vatikan_poradi").cast(int))
```

```python
vsepapezstvo = df_wikidata.join(df_vatikan, right_on="vatikan_jmeno", left_on="jmeno_en", how="full")
```

```python
vsepapezstvo.sort(by="poradi")
```

```python
vsepapezstvo_sparovane = dict(vsepapezstvo.filter(pl.col('jmeno_en') == pl.col('vatikan_jmeno')).select(pl.col(['id','vatikan_jmeno'])).rows_by_key(key='vatikan_jmeno'))
vsepapezstvo_sparovane = {jmeno: kod[0] for jmeno, kod in vsepapezstvo_sparovane.items()}
len(vsepapezstvo_sparovane)
```

```python
vsepapezstvo_sparovane
```

```python
vsepapezstvo.filter(pl.col('vatikan_poradi').is_null()).sort(by='pontifikat_zacatek')
```

```python
vsepapezstvo_nesparovane = vsepapezstvo.filter(pl.col('id').is_null()).select(pl.col('vatikan_jmeno','vatikan_pontifikat_zacatek'))

vsepapezstvo_nesparovane
```

```python
vsepapezstvo_nesparovane_doplneni = {
    'Francis':'Q450675',
    "Deusdedit or Adeodatus I":'Q170550',
    "Deusdedit or Adeodatus  I":'Q170550',
    "Benedict I":'Q151902',
    "Benedict  I":'Q151902',
    "Miltiades or Melchiades":'Q162090',
    "Eutichianus": 'Q162972',
    "Pontianus"	: "Q132113",
    "Callistus I":"Q122376",
    "Eleutherius":	"Q101090",
    "Soterus": "Q101280",
    "Clement": "Q42887",
    "Anacletus or Cletus":"Q80450",
    "Peter":"Q33923"
}
```

```python
vsepapezstvo_sparovane.update(vsepapezstvo_nesparovane_doplneni)
len(vsepapezstvo_sparovane)
```

```python
vsepapezstvo_sparovane
```

```python
len(df_vatikan)
```

```python
def dopln_kod(papez):
    
    for jmeno, kod in vsepapezstvo_sparovane.items():
        if papez == jmeno:
            return kod
```

```python
df_vatikan.with_columns(pl.col("vatikan_jmeno").map_elements(dopln_kod).alias("id")).sample(5)
```

```python
df_vatikan.with_columns(pl.col("vatikan_jmeno").map_elements(dopln_kod, return_dtype=str).alias("id")).filter(pl.col("id").is_null())
```

```python
df_vatikan = df_vatikan.with_columns(pl.col("vatikan_jmeno").map_elements(dopln_kod, return_dtype=str).alias("id"))
```

## Benedikte!

```python
from datetime import date
```

```python
df.filter(pl.col("vatikan_poradi").is_between(145,150))
```

```python
df = df.with_columns(
    pl.when(
        pl.col('vatikan_poradi') == 145
    ).then(
        date(year=1032, month=9, day=1)
    ).when(
        pl.col('vatikan_poradi') == 150
    ).then(
        date(year=1047, month=10, day=1)
    ).otherwise(
        pl.col('pontifikat_zacatek')
    ).alias('pontifikat_zacatek')
).with_columns(
    pl.when(
        pl.col('vatikan_poradi') == 145
    ).then(
        date(year=1044, month=9, day=1)
    ).when(
        pl.col('vatikan_poradi') == 150
    ).then(
        date(year=1048, month=8, day=1)
    ).otherwise(
        pl.col('pontifikat_konec')
    ).alias('pontifikat_konec')
)
```

```python
df.filter(pl.col("vatikan_jmeno") == "Benedict IX")
```
