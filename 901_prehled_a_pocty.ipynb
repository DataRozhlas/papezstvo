```python
import warnings
import polars as pl
import altair as alt

from src.kristi_promin import kristi_promin
from src.me_to_neurazi import me_to_neurazi
from src.alt_friendly import alt_friendly

alt.themes.register('irozhlas', kristi_promin)
alt.themes.enable('irozhlas')
warnings.filterwarnings('ignore')
```

```python
def stoleti(datum):
    return f"{int(str(datum)[0:2]) + 1}. století"
```

```python
df = pl.read_parquet('data/papezstvo.parquet', use_pyarrow=True).join(
    pl.read_parquet('data/svati_blahoslaveni.parquet', use_pyarrow=True), how='left', on='id'
).with_columns(
    (pl.col('pontifikat_zacatek').map_elements(stoleti).alias("století"))
).with_columns(
    pl.col("století").rank("ordinal").over("století").alias("poradi_ve_stoleti")
)
```

```python
poradi_stoleti = df.select(pl.col("století")).to_series().to_list()
```

```python
historicky_prehled = alt.Chart(
    df.to_pandas(),
    title=alt.Title(
        "Přibližné počty papežů a jejich kanonizace",
        subtitle=["Zlatě jsou vybarveni papežové svatořečení, červeně blahoslavení.",
                  "Papeže úřadující na přelomech století graf zobrazuje pouze v tom, kdy byli zvoleni.",
                 "Zato Benedikt IX. je v 11. století znázorněn třemi body – vládl ve třech obdobích."]
    )
).mark_point(filled=True).encode(
    alt.X("století:N", title=None, sort=poradi_stoleti),
    alt.Y("poradi_ve_stoleti", title=None),
    alt.Color("status:N", scale=alt.Scale(range=['gray','#B72D0D','#E99A2B']), legend=None),
    alt.Opacity("status:N", scale=alt.Scale(range=[0.5,1,0.5])),
    ).configure_view(
    stroke='transparent'
).configure_axis(grid=True, gridOpacity=0.6, gridWidth = 0.5, gridColor='#F1D1BB').properties(
    width=380,
    height=180
)

historicky_prehled
```
