```python
def proklepni_papeze(papez):
    
    data_o_papezi = {}
    
    try:
        data_o_papezi['jmeno'] = papez['entities'][list(papez['entities'])[0]]['labels']['cs']['value']
    except:
        data_o_papezi['jmeno'] = papez['entities'][list(papez['entities'])[0]]['labels']['en']['value']

    try:
        data_o_papezi['jmeno_en'] = papez['entities'][list(papez['entities'])[0]]['labels']['en']['value']
    except:
        pass
    
    try:
        data_o_papezi['narozeni_misto'] = papez['entities'][list(papez['entities'])[0]]['claims']['P19'][0]['mainsnak']['datavalue']['value']['id']
    except:
        data_o_papezi['narozeni_misto'] = None

    try:
        data_o_papezi['narozeni_datum'] = papez['entities'][list(papez['entities'])[0]]['claims']['P569'][0]['mainsnak']['datavalue']['value']['time'][1:11]
    except:
        data_o_papezi['narozeni_datum'] = None

    try:
        data_o_papezi['narozeni_presnost'] = papez['entities'][list(papez['entities'])[0]]['claims']['P569'][0]['mainsnak']['datavalue']['value']['precision']
    except:
        data_o_papezi['narozeni_presnost'] = None

    try:
        data_o_papezi['krest_datum'] = papez['entities'][list(papez['entities'])[0]]['claims']['P1636'][0]['mainsnak']['datavalue']['value']['time'][1:11]
    except:
        data_o_papezi['krest_datum'] = None

    try:
        data_o_papezi['umrti_misto'] = papez['entities'][list(papez['entities'])[0]]['claims']['P20'][0]['mainsnak']['datavalue']['value']['id']
    except:
        data_o_papezi['umrti_misto'] = None

    try:
        data_o_papezi['umrti_datum'] = papez['entities'][list(papez['entities'])[0]]['claims']['P570'][0]['mainsnak']['datavalue']['value']['time'][1:11]
    except:
        data_o_papezi['umrti_datum'] = None

    try:
        data_o_papezi['umrti_presnost'] = papez['entities'][list(papez['entities'])[0]]['claims']['P570'][0]['mainsnak']['datavalue']['value']['precision']
    except:
        data_o_papezi['umrti_presnost'] = None

    try:
        data_o_papezi['umrti_kategorie'] = papez['entities'][list(papez['entities'])[0]]['claims']['P1196'][0]['mainsnak']['datavalue']['value']['id']
    except:
        data_o_papezi['umrti_kategorie'] = None

    try:
        data_o_papezi['umrti_pricina'] = papez['entities'][list(papez['entities'])[0]]['claims']['P509'][0]['mainsnak']['datavalue']['value']['id']
    except:
        data_o_papezi['umrti_pricina'] = None

    try:
        for funkce in papez['entities'][list(papez['entities'])[0]]['claims']['P39']:
            if funkce['mainsnak']['datavalue']['value']['id'] == 'Q172748':
                data_o_papezi['pontifikat_zacatek'] = funkce['qualifiers']['P580'][0]['datavalue']['value']['time'][1:11]
    except:
        data_o_papezi['pontifikat_zacatek'] = None
    
    try:
        for funkce in papez['entities'][list(papez['entities'])[0]]['claims']['P39']:
            if funkce['mainsnak']['datavalue']['value']['id'] == 'Q172748':
                data_o_papezi['pontifikat_konec']  = funkce['qualifiers']['P582'][0]['datavalue']['value']['time'][1:11]
    except:
        data_o_papezi['pontifikat_konec'] = None
    
    try:
        rodne_jazyky = []
        for rodny_jazyk in papez['entities'][list(papez['entities'])[0]]['claims']['P103']:
            rodne_jazyky.append(rodny_jazyk['mainsnak']['datavalue']['value']['id'])
        data_o_papezi['rodny_jazyk'] = rodne_jazyky
    except:
        data_o_papezi['rodny_jazyk'] = None

    try:
        dalsi_jazyky = []
        for dalsi_jazyk in papez['entities'][list(papez['entities'])[0]]['claims']['P1412']:
            dalsi_jazyky.append(dalsi_jazyk['mainsnak']['datavalue']['value']['id'])
        data_o_papezi['dalsi_jazyky'] = dalsi_jazyky
    except:
        data_o_papezi['dalsi_jazyky'] = None

    try:
        povolani = []
        for profese in papez['entities'][list(papez['entities'])[0]]['claims']['P106']:
            povolani.append(profese['mainsnak']['datavalue']['value']['id'])
        data_o_papezi['povolani'] = povolani
    except:
        data_o_papezi['povolani'] = None

    try:
        rady = []
        for rad in papez['entities'][list(papez['entities'])[0]]['claims']['P611']:
            rady.apppend(rad['mainsnak']['datavalue']['value']['id'])
        data_o_papezi['rady'] = rady
    except:
        data_o_papezi['rady'] = None

    try:
        ucastnil_se = []
        for ucastnik in papez['entities'][list(papez['entities'])[0]]['claims']['P1344']:
            ucastnil_se.append(ucastnik['mainsnak']['datavalue']['value']['id'])
        data_o_papezi['ucastnik'] = ucastnil_se
    except:
        data_o_papezi['ucastnik'] = None

    data_o_papezi['wikihesla'] = [x for x in list(papez['entities'][list(papez['entities'])[0]]['sitelinks'].keys()) if x[-4:] == 'wiki']
    
    return data_o_papezi

papezstvo = []

for p in os.listdir("downloads/antipopes"):
    if 'Q' in p:
        with open(os.path.join(cesta, p), 'r', encoding='utf-8') as soubor_s_papezem:
            obsah_souboru_s_papezem = json.loads(soubor_s_papezem.read())
            papezstvo.append(proklepni_papeze(obsah_souboru_s_papezem))
```

```python
df = pl.DataFrame(papezstvo, orient="row", infer_schema_length=None)
```
