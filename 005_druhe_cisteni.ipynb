```python
import os
import json
import polars as pl
```

```python
pl.Config(tbl_rows=30)
```

```python
odkud = "downloads/wikidata_labels"

stazene = [f for f in os.listdir(odkud)]

slovnik = {}
jazyky = 'cs|en|sk|de|fr|es|ru'.split('|')
for f in stazene:
    with open(os.path.join(odkud, f), 'r', encoding='utf-8') as soubor:
        minislovnik = json.loads(soubor.read())
        try:
            for entita in minislovnik['entities']:
                labels = minislovnik['entities'][entita].get('labels')
                value = None
                for j in jazyky:
                    if value == None:
                        try:
                            value = labels[j]['value']
                        except:
                            pass
                slovnik[entita] = value
        except:
            pass
```

```python
slovnik
```

```python
df = pl.read_parquet("data_raw/papezstvo_raw.parquet", use_pyarrow=True)
```

```python
df
```

```python
# For the list columns, we need to "explode" them, do the replacement, and then collect them back
list_cols = [col for col in df.columns if df.schema[col] == pl.List]
string_cols = [col for col in df.columns if df.schema[col] == pl.String]
print(f"seznamy: {list_cols}")
print(f"ostatní: {string_cols}")

df = df.with_columns(pl.col('narozeni_misto').alias('narozeni_misto_kod'))

# Replace in non-list columns normally
df = df.with_columns(pl.col(string_cols).replace(slovnik))

# For list columns, we need to apply the replacement to the list elements
for col in list_cols:
    df = df.with_columns(
        pl.col(col).map_elements(lambda x: [slovnik.get(item, item) for item in x])
    )
```
